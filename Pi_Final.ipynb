```python
# Archivo user_reviews.json.gz
import pandas as pd
import gzip
import json
import ast
```

```python
 #leer el archivo user_reviews.json.gz comprimid
def leer_user_r():
    with gzip.open('user_reviews.json.gz','rt',encoding='utf-8')as file:
        return[ast.literal_eval(line.strip())
               for line in file]

datos= leer_user_r()

#convertirlo a un dataframe
df=pd.DataFrame(datos)
df.head()
```

```python
# Archivo steam_games.json.gz
import pandas as pd
import gzip
import json

def leer_user_g():
    with gzip.open('steam_games.json.gz', 'rt', encoding='utf-8') as file:
        data = [json.loads(line.strip()) for line in file]
    return data

datos_g = leer_user_g()

# Convertirlo a un DataFrame
df3 = pd.DataFrame(datos_g)
df3.head()
```
